### Przyk≈Çadowy program w Jupyter

```python
#importowanie bibliotek
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('D:/DYDAKTYKA/Python/dane/Si.txt', skiprows=2) 
print(data[0:5,:])
```

```python
energy = data[:,0]
print(f'Energia fotonu {energy[:5]}')
n = data[:,1]
print(f'n {energy[:5]}')
k = data[:,2]
print(f'k: {energy[:5]}')
```

#### Wizualizacja danych 

```python
plt.plot(energy, n)
plt.plot(energy, k)
plt.xlabel('Energia fotonu, eV')
plt.ylabel('n, k')
plt.show()
```

#### Przeliczenie z eV --> nm 
## $nm = \frac{1026}{eV}$

```python
wvl = 1026/energy
print(wvl[:5])
```

```python
plt.plot(wvl, n)
plt.plot(wvl, k)
plt.xlabel('wavelength, nm')
plt.ylabel('n, k')
plt.show()
```
